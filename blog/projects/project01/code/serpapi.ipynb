Serp apiを使用して検索する
アカウントを作成して、ページにログインすると、APIKEYをコピーできるようになります。
そのAPIKEYを、「.env」ファイルの中に設定します。
（.env.exampleを参照して作成します。）

```python
%pip install python-dotenv requests
```

```python
from dotenv import load_dotenv
import requests
import json
import os

load_dotenv()

def get_book_price_serpapi(book_title):
    api_key = os.getenv("SERPAPI_API_KEY")
    if not api_key:
        print('".env" を設定してください。".env.example"を参照')
        return

    search_engine_id = "google"
    
    # 使用する検索エンジンとパラメータを設定
    params = {
        'api_key': api_key,
        'engine': search_engine_id,
        'q': f'{book_title} 中古 価格',
        'gl': 'jp',  # 国コードを日本に設定
        'lr': 'lang_ja',  # 言語を日本語に設定
    }
    
    url = "https://serpapi.com/search.json"
    response = requests.get(url, params=params)
    data = json.loads(response.text)
    
    if 'organic_results' in data:
        for result in data['organic_results']:
            title = result.get('title', '')
            link = result.get('link', '')
            snippet = result.get('snippet', '')
            
            # 結果を表示
            print(f"Title: {title}")
            print(f"Link: {link}")
            print(f"Snippet: {snippet}\n")
    else:
        print("No results found.")

if __name__ == "__main__":
    book_title = input("本のタイトルを入力してください: ")
    get_book_price_serpapi(book_title)
```
