VALUE SERPを使用して検索する
アカウントを作成して、ページにログインすると、APIKEYをコピーできるようになります。
そのAPIKEYを、「.env」ファイルの中に設定します。
（.env.exampleを参照して作成します。）

```python
from pathlib import Path
from dotenv import load_dotenv
import requests
import json
import os

# このノートブックと同じ code/ にある .env を読む
load_dotenv(Path.cwd() / ".env")
load_dotenv(Path.cwd() / "blog/projects/project01/code/.env")

def get_book_price_valueserp(book_title):
    api_key = os.getenv("VALUESERP_API_KEY")
    if not api_key:
        print('".env" を設定してください。".env.example"を参照')
        return

    params = {
        'api_key': api_key,
        'q': f'{book_title} 中古 価格',
        'gl': 'jp',
        'hl': 'ja',
        'google_domain': 'google.co.jp',
    }

    url = "https://api.valueserp.com/search"
    response = requests.get(url, params=params)
    data = response.json()

    organic_results = data.get('organic_results', [])
    if not organic_results:
        print("No results found.")
        return []

    for result in organic_results:
        title = result.get('title', '')
        link = result.get('link', '')
        snippet = result.get('snippet', '')

        print(f"Title: {title}")
        print(f"Link: {link}")
        print(f"Snippet: {snippet}\n")

    return organic_results
```

```python
# 実行セル（input は使わない。タイトルを書き換えて実行）
book_title = "新詳説DTP基礎"
get_book_price_valueserp(book_title)
```
